## What is Data Visualization?

- **Definition**: The practice of translating information into a visual context  
- **Purpose**:
  - Simplify the interpretation of complex data  
  - Reveal patterns, trends, and correlations  
  - Enhance data-driven decision making

## Example Activity:
  - Plot simple line charts and bar charts using sample code snippets

```python
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
import numpy as np

# Sample data
data = [1, 3, 2, 5, 7, 4]

# Matplotlib example
# Matplotlib works best with numpy arrays
plt.plot(data)
plt.title("Simple Line Chart")
plt.show()


# Define a dataframe:
df = pd.DataFrame({
    'x': [chr(97 + i) for i in range(len(data))],
    'y': data
})


# Seaborn example
# Seaborn provides better support for pandas DataFrames
# Create a bar plot with categories `on the x-axis and values on the y-axis
sns.barplot(data=df, x='x', y='y')

plt.xticks()  # Rotate x-axis labels for better readability
plt.show()
```

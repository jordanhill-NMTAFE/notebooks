```python
import numpy as np
from sklearn import datasets
from sklearn.semi_supervised import LabelSpreading
import matplotlib.pyplot as plt
```

```python
# Create a synthetic dataset
X, y = datasets.make_moons(n_samples=1000, noise=0.05)
```

```python
plt.figure(figsize=(10, 8))
colors = {-1: 'gray', 0: 'red', 1: 'blue'}
labels = {-1: 'Unlabeled', 0: 'Class 0', 1: 'Class 1'}

for label in colors:
    indices = (y == label)
    plt.scatter(X[indices, 0], X[indices, 1], c=colors[label], label=labels[label], edgecolors='k', s=60)

# Plot the predicted labels
# plt.scatter(X[:, 0], X[:, 1], c=[colors[label] for label in y_pred], edgecolors='k', alpha=1, s=60)

# Add a legend
plt.legend(loc='best')
plt.title("Semi-Supervised Learning with Label Spreading")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python

y[100:] = -1  # Unlabel majority of the data
# y[0:100] = [0] * 50 + [1] * 50  # Label a small subset
```

```python
plt.figure(figsize=(10, 8))
colors = {-1: 'gray', 0: 'red', 1: 'blue'}
labels = {-1: 'Unlabeled', 0: 'Class 0', 1: 'Class 1'}

for label in colors:
    indices = (y == label)
    plt.scatter(X[indices, 0], X[indices, 1], c=colors[label], label=labels[label], edgecolors='k', s=60)

# Plot the predicted labels
# plt.scatter(X[:, 0], X[:, 1], c=[colors[label] for label in y_pred], edgecolors='k', alpha=1, s=60)

# Add a legend
plt.legend(loc='best')
plt.title("Semi-Supervised Learning with Label Spreading")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
# Create and fit the model
label_spread = LabelSpreading(kernel='knn', alpha=0.3)
label_spread.fit(X, y)

# Predict labels for the entire dataset
y_pred = label_spread.transduction_
```

```python
# Plot the results
plt.figure(figsize=(10, 8))
colors = {-1: 'gray', 0: 'red', 1: 'blue'}
labels = {-1: 'Unlabeled', 0: 'Class 0', 1: 'Class 1'}

for label in colors:
    indices = (y_pred == label)
    plt.scatter(X[indices, 0], X[indices, 1], c=colors[label], label=labels[label], edgecolors='k', s=60)

# Plot the predicted labels
# plt.scatter(X[:, 0], X[:, 1], c=[colors[label] for label in y_pred], edgecolors='k', alpha=1, s=60)

# Add a legend
plt.legend(loc='best')
plt.title("Semi-Supervised Learning with Label Spreading")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
